Bernoulli Distribution: Describes the probability distribution of a random variable that takes the value 1 with probability

Binomial Distribution: Describes the number of successes in a fixed number of independent Bernoulli trials.
Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given a constant average rate of occurrence.
Continuous Probability Distributions:

Uniform Distribution: All outcomes are equally likely within a given range.
Normal (Gaussian) Distribution: Describes a symmetric, bell-shaped curve, characterized by its mean and standard deviation. It is widely used due to the Central Limit Theorem.
Exponential Distribution: Models the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.

```python
from scipy.stats import norm

def calculate_normal_pdf(mean, std_dev, x):
    """
    Calculate the probability density function (PDF) of a normal distribution
    at a given point x.

    Parameters:
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.
        x (float): Point at which to evaluate the PDF.

    Returns:
        pdf (float): Probability density function value at point x.
    """
    pdf = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf

# Example usage:
mean = 0
std_dev = 1
x = 1.5
pdf_value = calculate_normal_pdf(mean, std_dev, x)
print("PDF at x =", x, ":", pdf_value)
```

The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. The properties of the binomial distribution are as follows:

Fixed Number of Trials: The number of trials, denoted by 
�
n, is fixed in advance.
Independent Trials: Each trial is independent of the others.
Two Possible Outcomes: Each trial has only two possible outcomes: success (usually denoted by 
�
p) or failure (denoted by 
1
−
�
1−p).
Constant Probability of Success: The probability of success, denoted by 
�
p, remains constant from trial to trial.
Discrete Random Variable: The binomial random variable represents the count of successes in the 
�
n trials.
Examples of events where the binomial distribution can be applied:

Coin Flips: Consider flipping a fair coin 
�
n times. Each flip is a Bernoulli trial, where success might represent getting heads and failure represents getting tails. The number of heads obtained in 
�
n flips follows a binomial distribution.
Manufacturing Defects: In a manufacturing process, each item produced might be defective (failure) or non-defective (success) with a certain probability. The number of defective items in a batch of 
�
n items can be modeled using a binomial distribution, assuming that the items are produced independently and with the same probability of being defective.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
probability_of_success = 0.4
sample_size = 1000

# Generate random sample from binomial distribution
random_sample = np.random.binomial(1, probability_of_success, sample_size)

# Plot histogram
plt.hist(random_sample, bins=[-0.5, 0.5, 1.5], edgecolor='black')  # Binomial has only two possible outcomes (0 or 1)
plt.title("Histogram of Binomial Distribution")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.xticks([0, 1], ['Failure', 'Success'])  # Label the x-axis with 'Failure' and 'Success'
plt.grid(True)
plt.show()
```

```python
from scipy.stats import poisson

def calculate_poisson_cdf(mean, x):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution
    with the given mean at a given point.

    Parameters:
        mean (float): Mean of the Poisson distribution.
        x (int): Point at which to evaluate the CDF.

    Returns:
        cdf (float): Cumulative distribution function value at point x.
    """
    cdf = poisson.cdf(x, mu=mean)
    return cdf

# Example usage:
mean = 3
x = 2
cdf_value = calculate_poisson_cdf(mean, x)
print("CDF at x =", x, ":", cdf_value)
```
